```python
def taxicab_similarity(corpus_list):
    all_keys = []
    final_distances = []
    for book in corpus_list:
        for song in book.values():
            song_fingering = []
            song_fingering_fixed = []
            song_fingering_distances = []
            all_keys.append((song['data']['key'], song['data']['final']))
            for x in song['alfabeto']:
                if x in AlfabetoDict.keys():
                    song_fingering.append(AlfabetoDict[x])
                elif type(x) is int:
                    song_fingering.append(AlfabetoDict[x])
            song_fingering_fixed.append(song_fingering[-1])
            for x in range(1, len(song_fingering)):
                if song_fingering[x-1] in AlfabetoDict.values() and type(song_fingering[x]) is int:
                    song_fingering_fixed.append([y + song_fingering[x] for y in song_fingering[x-1]])
                elif song_fingering[x] != song_fingering[x-1]:
                    song_fingering_fixed.append(song_fingering[x])
            for x in range(1, len(song_fingering_fixed)):
                song_fingering_distances.append(cityblock(song_fingering_fixed[x-1], song_fingering_fixed[x]))
            final_distances.append(np.mean(song_fingering_distances))
    return final_distances, all_keys
```

```python
def taxicab_averages(taxicab_sim_data):
    all_keys = []
    all_dist = {}
    for x in taxicab_sim_data[1]:
        if x not in all_keys:
            all_keys.append(x)
    for i in all_keys:
        temp_list = []
        for x, y in zip (taxicab_sim_data[0], taxicab_sim_data[1]):
            if i == y:
                temp_list.append(x)
        all_dist[i] = np.mean(temp_list)
    return all_dist
```

```python
def number_of_fingers(corpus_list):
    all_keys = []
    final_fingering_percents = []
    for book in corpus_list:
        for song in book.values():
            song_fingering = []
            song_fingering_fixed = []
            total_finger_numbers = []
            fingering_percent = []
            all_keys.append((song['data']['key'], song['data']['final']))
            for x in song['alfabeto']:
                if x in AlfabetoDict.keys():
                    song_fingering.append(AlfabetoDict[x])
                elif type(x) is int:
                    song_fingering.append(AlfabetoDict[x])
            song_fingering_fixed.append(song_fingering[-1])
            for x in range(1, len(song_fingering)):
                if song_fingering[x-1] in AlfabetoDict.values() and type(song_fingering[x]) is int:
                    song_fingering_fixed.append([y + song_fingering[x] for y in song_fingering[x-1]])
                elif song_fingering[x] != song_fingering[x-1]:
                    song_fingering_fixed.append(song_fingering[x])
            for x in song_fingering_fixed:
                number = 0
                for y in x:
                    if y > 0:
                        number += 1
                total_finger_numbers.append(number)
#                 total_finger_numbers.append(sum(x))
            final_fingering_percents.append(np.mean(total_finger_numbers))
    return final_fingering_percents, all_keys
```
