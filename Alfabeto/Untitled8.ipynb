```python
n_neighbors = 6

# import some data to play with
iris = datasets.load_iris()
# X = iris.data[:, :2]  # we only take the first two features. We could
                      # avoid this ugly slicing by using a two-dim dataset
XX = np.array(di.alfabeto_notes_data[0])
X = PCA(n_components=2).fit_transform(XX)

lab_list = []
# for mode in di.zmo_notes_data[1]:
#     if mode in di.modal_major:
#         lab_list.append(0)
#     elif mode in di.modal_minor:
#         lab_list.append(1)
#     else:
#         lab_list.append(2)
for mode in di.alfabeto_notes_data[1]:
    if mode in di.modal_types['ionian']:
        lab_list.append(0)
    elif mode in di.modal_types['dorian']:
        lab_list.append(1)
    elif mode in di.modal_types['phrygian']:
        lab_list.append(2)
    elif mode in di.modal_types['lydian']:
        lab_list.append(3)
    elif mode in di.modal_types['mixolydian']:
        lab_list.append(4)
    elif mode in di.modal_types['aeolian']:
        lab_list.append(5)
    else:
        lab_list.append(6)
y = np.array(lab_list)    
# y = iris.target

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF', '#FF0000', '#00FF00', '#0000FF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF', '#FFAAAA', '#AAFFAA', '#AAAAFF'])

for weights in ['uniform', 'distance']:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))

plt.show()
```

```python
import random, copy
from timeit import timeit
from music21 import *


class CounterPoint:
    perfect_intervals = (0, 7, 12)
    illegal_intervals = (1, 2, 5, 6, 10, 11)
    thirds = (3, 4)
    sixths = (8, 9)

    diatonic_pcs = (0, 2, 4, 5, 7, 9, 11)
    chromatic_pcs = (1, 3, 6, 8, 10)

    illegal_skips = range(9, 12)


    pc_modes = {
        'ionian': (0, (11, 2)),
        'dorian': (2, (1, 4)),
        'phrygian': (4, (2, 5)),
        'lydian': (5, (4, 7)),
        'mixolydian': (7, (6, 9)),
        'aeolian': (9, (8, 11))
    }

    clefs = {'treble': range(57, 79), 'bass': range(45, 62)}

    # def __init__(self, mode='ionian', length=6, clef='treble'):
    #     self.mode = mode
    #     self.length = length
    #     self.clef = clef

    def melody_maker(self, clef='treble', length=6, mode='ionian'):
        # clef = self.clef
        # length = self.length
        # mode = self.mode
        '''find all possible notes'''
        mel = [0 for x in range(length)]
        possible_notes = []
        possible_cadences = []
        for x in self.clefs[clef]:
            if x%12 in self.diatonic_pcs:
                possible_notes.append(x)
            if x%12 in self.pc_modes[mode][1]:
                possible_cadences.append(x)
        possible_notes = random.sample(possible_notes, len(possible_notes))

        '''make the cadence'''
        possible_cadences = random.sample(possible_cadences, len(possible_cadences))
        possible_final = []
        for x in possible_notes:
            if x%12 == self.pc_modes[mode][0]:
                possible_final.append(x)
        thefinal = random.choice(possible_final)
        mel[0] = thefinal
        mel[-1] = thefinal
        mel[-2] = random.choice(possible_cadences)

        for n in range(1, len(mel[:-2])):
            next_note = []
            for nn in possible_notes:
                if nn != mel[n-1]: #no repeated notes
                    if abs(nn - mel[n-1]) not in self.illegal_skips: #no illegal skips
                        if n == len(mel[1:-2]) and mel[-2] in self.chromatic_pcs: #check for augmented second
                            if mel[-2] - n < 3: #and leap into chromatic note
                                next_note.append(nn)
                        else:
                            next_note.append(nn)
            mel[n]=random.choice(next_note)
        test_mel = self.check_melody(mel=mel, mode=mode)
        while test_mel is False:
            
            mel = self.melody_maker(clef=clef, length=length, mode=mode)
            print(mel, 'melody for testing')
            test_mel = self.check_melody(mel=mel, mode=mode)
        return mel

    def check_melody(self, mel, mode, zenith_threshold=1, nadir_threshold=1):
        if mel[0] != mel[-1]:
            print('it was the first and last note')
            return False
        elif mel[-2]%12 not in self.pc_modes[mode][-1]:
            print('penultimate')
            return False
        for x in range(1, len(mel)):
            if mel[x - 1] == mel[x]:
                print('consecutive', mel[x-1], mel[x])
                return False

            interval = abs(mel[x] - mel[x - 1])

            if interval in range(10, 12) or interval > 12:
                print('leap too big', mel[x], mel[x-1], interval)
                return False
            elif interval == 6:
                print('tritone')
                return False
            elif mel[x] % 12 not in self.diatonic_pcs and mel[x] - mel[x - 1] == 3:
                # print(mel[x]%12, mel[x-1]%12, mel[x] - mel[x-1])
                print('augmented second')
                return False
            elif mel[x] % 12 not in self.diatonic_pcs and abs(mel[x] - mel[x - 1]) % 12 > 4:
                print('leap into dissonance', mel[x] % 12, mel[x - 1] % 12, abs(mel[x] - mel[x - 1]) % 12)
                return False
        zenith = 0
        nadir = 0
        for x in mel[1:-1]:
            if x == max(mel):
                zenith += 1
            elif x == min(mel):
                nadir += 1
        if nadir > nadir_threshold:
            print('it was the nadir', min(mel), nadir)
            return False
        if zenith > zenith_threshold:
            print('it was the zenith', max(mel), zenith)
            return False
        # self.melody.append(mel)
        return True

    def counterpoint_checker(self, cf, cpt):
        all_intervals = []
        for n in range(len(cpt) - 1):
            interval = cpt[n] - cf[n]
            all_intervals.append(interval)
            p_interval = cpt[n + 1] - cf[n + 1]
            if interval >= 20 or interval < 0:  # range issues
                print('range issues', interval)
                return False
            elif interval % 12 in self.illegal_intervals:  # illegal harmonic interval
                print('illegal interval', interval, interval % 12)
                return False
            elif interval == 7 and p_interval == 7:  # parallel fifths
                print('consecutive fifths', interval, p_interval)
                return False
            elif p_interval in self.perfect_intervals and cpt[n] - cpt[n - 1] > 2 and cf[n] - cf[n - 1] > 2:
                print('direct perfects')
                return False
            elif p_interval in self.perfect_intervals and cpt[n - 1] - cpt[n] > 2 and cf[n - 1] - cf[n] > 2:
                print('direct perfects')
                return False
        print('checking thirds')
        thirds_counter = 0
        sixths_counter = 0
        for n in range(len(cpt)):
            interval = (cpt[n] - cf[n]) % 12
            if interval in self.thirds:
                thirds_counter += 1
                if thirds_counter >= 4:
                    print('too many thirds')
                    return False
            else:
                thirds_counter = 0
            if interval in self.sixths:
                sixths_counter += 1
                if sixths_counter >= 4:
                    print('too many sixths')
                    return False
            else:
                sixths_counter = 0
        upward_motion = 0
        downward_motion = 0
        for n in range(1, len(cpt)):
            if cpt[n] > cpt[n - 1] and cf[n] > cf[n - 1]:  # no contrary motion
                upward_motion += 1
                downward_motion = 0
            if upward_motion > 3:
                print('upward motion')
                return False
            elif cpt[n] < cpt[n - 1] and cf[n] < cf[n - 1]:
                upward_motion = 0
                downward_motion += 1
            if downward_motion > 3:
                print('downward motion')
                return False
        for n in range(1, len(cpt)):  # leap in same direction
            if cpt[n - 1] - cpt[n] > 3 and cf[n - 1] - cf[n] > 3:
                print('leap same direction')
                return False
            elif cpt[n] - cpt[n - 1] > 3 and cf[n] - cf[n - 1] > 3:
                print('leap same direction')
                return False
        if abs(cf[-2] - cf[-1]) > 2 and abs(cpt[-2] - cpt[-1]) > 2:  # cadence
            print('cadence intervals')
            return False
        elif abs(cf[-2] - cf[-1]) > 7 or abs(cpt[-2] - cpt[-1]) > 7:
            print('cadence intervals')
            return False
        elif cf[-2] % 12 in self.chromatic_pcs and cpt[-3] % 12 == cf[-2] % 12 - 1:  # cross relations
            print('cross relation')
            return False
        elif cpt[-2] % 12 in self.chromatic_pcs and cf[-3] % 12 == cpt[-2] % 12 - 1:
            print('cross relation')
            return False
        for a, b in zip(cf[1:-1], cpt[1:-1]):
            if abs(a - b) % 12 == 0:
                print('octave in middle')
                return False
        return True

    def make_counterpoint(self, length=6, mode='ionian', cf_clef='bass', cpt_clef='treble'):
        cpt = self.melody_maker(length=length, mode=mode, clef=cpt_clef)
        cf = self.melody_maker(length=length, mode=mode, clef=cf_clef)
        if len(cf) != len(cpt):
            raise ValueError('Cantus and Counterpoint melodies must be the same length')
        trial_counterpoint = self.counterpoint_checker(cf=cf, cpt=cpt)
        while trial_counterpoint is False:
            cpt = self.melody_maker(length=length, mode=mode, clef=cpt_clef)
            cf = self.melody_maker(length=length, mode=mode, clef=cf_clef)
            trial_counterpoint = self.counterpoint_checker(cf=cf, cpt=cpt)
        return {'counterpoint': cpt, 'cantus': cf}

# def cf_maker(mode, clef, length):
#     '''makes the cantus firmus'''
#
#     '''final cadence'''
#     final_candidates = []
#     for b_note in clef:
#         if b_note % 12 == pc_modes[mode][0]:
#             final_candidates.append(b_note)
#     # print(final_candidates)
#     final = random.choice(final_candidates)
#
#     '''make cadence'''
#     cf = [0 for x in range(length)]
#     cadence = pc_modes[mode][-1]
#
#     cf[0] = final  # first note of cf
#     cf[-1] = final  # last note of cf
#     cadence_choices = []
#     for c_note in clef:
#         if c_note % 12 in cadence:
#             cadence_choices.append(c_note)
#     cf[-2] = random.choice(cadence_choices)  # penultimate note of cf
#     # print(cf)
#
#     '''make the rest'''
#     '''diatonic, skips (don't repeat, under ic 8, no ic 6, no +2), '''
#
#     def try_cf(cf):
#         cf_tester = copy.deepcopy(cf)
#         note_choices = []  # ensures diatonic
#         for note_choice in clef:
#             if note_choice % 12 in diatonic_pcs:
#                 note_choices.append(note_choice)
#         for n in range(1, length - 2):
#             next_note_choices = []
#             for nn in note_choices:
#                 if nn != cf_tester[n - 1] and abs(nn - cf_tester[n - 1]) % 12 != 6:
#                     if abs(nn-cf_tester[n-1]) <= 9 or abs(nn-cf_tester[n-1])%12 == 0:
#
#                         next_note_choices.append(nn)
#             cf_tester[n] = random.choice(next_note_choices)
#         # print([x%12 for x in cf_tester])
#         return cf_tester
#
#     C = CptCheck()
#     cf_attempt = try_cf(cf)
#     attempt = 0
#     tester = C.check_melody(cf_attempt, mode)
#     while tester is False:
#         cf_attempt = try_cf(cf)
#         tester = C.check_melody(cf_attempt, mode)
#         attempt += 1
#         print([x % 12 for x in cf_attempt])
#         print(cf_attempt, 'midi cf attempt')
#         if attempt > 1000:
#             print('tried 1000 times, new cf')
#             print('this one was bad', cf)
#             cf_maker(mode, clef, length)
#
#     return cf_attempt
#     # def check_form(cf_tester):
#     #     cf_tester = try_cf(cf_tester)
#     #     # print('for testing', cf_tester)
#     #     zenith = max(cf_tester)
#     #     high_notes = 0
#     #     nadir = min(cf_tester)
#     #     low_notes = 0
#     #     for x in cf_tester[1:-1]:
#     #         if x == zenith:
#     #             high_notes += 1
#     #         elif x == nadir:
#     #             low_notes += 1
#     #     if high_notes >= 2 or low_notes >= 2:
#     #         # print('trying again', cf)
#     #         check_form(try_cf(cf))
#     #     else:
#     #         print(cf_tester)
#     #         return cf_tester
#
#     # return check_form(try_cf(cf))
#
#
# def counterpoint_maker(species, cf, clef, mode, length):
#     cpt = [0 for x in range(len(cf))]
#     note_options = []
#     for c_note in clef:
#         if c_note % 12 in diatonic_pcs:
#             note_options.append(c_note)
#     for s_note in note_options:
#         if s_note == cf[0] or s_note == cf[0] + 12:
#             cpt[0] = s_note
#             cpt[-1] = s_note
#
#
#     # print('cpt testing...', cpt)
#
#     def cpt_builder():
#         perfect_intervals = (0, 7, 12)
#         illegal_intervals = (0, 1, 2, 5, 6, 10, 11)
#         # print(len(cpt[1:-1]))
#         cadences = []
#
#         for a in range(1, len(cpt[:-1])):
#             # print('made it?')
#             note_possibilities = []
#             for b in note_options:
#                 if b >= cf[a] and (b - cf[a]) % 12 not in illegal_intervals:
#                     if b != cpt[a-1] and abs(b-cpt[a-1]) < 17:
#
#                         note_possibilities.append(b)
#                     # print('good so far 1')
#                     #       note_possibilities.append(b)
#                     # print('good so far 2')
#                     #     note_possibilities.append(b)
#             if len(note_possibilities) >= 1:
#
#                 cpt[a] = random.choice(note_possibilities)
#             else:
#                 print('no note possibilities?')
#                 cpt_builder()
#         counterpoint = copy.deepcopy(cpt)
#
#         C = CptCheck()
#
#         cpt_attempt = C.check_melody(counterpoint, mode)
#         mel_att = 0
#         # print('checking notes')
#         print(counterpoint, 'counterpoint')
#         print(cf, 'cantus')
#         while cpt_attempt is False:
#             print(counterpoint, 'counterpoint did not work')
#             mel_att += 1
#             counterpoint = cpt_builder()
#             cpt_attempt = C.check_melody(counterpoint, mode)
#         return counterpoint
#
#     C = CptCheck()
#     print('made it to counterpoint--------------------------------------------')
#     print('cantus', cf)
#     cpt_try = cpt_builder()
#     cpt_tester = C.check_counterpoint(cf, cpt_try)
#     final_try = cf, cpt_try
#     c_att = 0
#     while cpt_tester is False:
#         c_att += 1
#         cpt_try = cpt_builder()
#         cpt_tester = C.check_counterpoint(cf, cpt_try)
#         final_try = cf, cpt_try
#     print('WE MADE IT!!!')
#     print([x % 12 for x in cpt_try])
#     print([x % 12 for x in cf])
#     return final_try


from time import time

st = time()
notes = {0: 'C', 1: 'C#', 2: 'D', 3: 'D#', 4: 'E', 5: 'F', 6: 'F#', 7: 'G', 8: 'G#', 9: 'A', 10: 'B-', 11: 'B'}
octaves = {'2': [x for x in range(36, 36+12)],
           '3': [x for x in range(48, 48 + 12)],
           '4': [x for x in range(60, 60 + 12)],
           '5': [x for x in range(72, 72 + 12)]}


# cantus_f = cf_maker(mode, bass_range, length)
# ctpt = counterpoint_maker('first', cantus_f, treble_range, mode, length)
cantus = stream.Part(id='cantus')
counterpoint = stream.Part(id='counterpoint')
C = CounterPoint()
# ctpt = C.melody_maker(clef='treble', length=6, mode='ionian')
# cf = C.melody_maker(clef='bass', length=6, mode='ionian')
final_cpt = C.make_counterpoint(mode='phrygian', cf_clef='bass', cpt_clef='treble')

print(final_cpt['counterpoint'], 'counterpoint')
print(final_cpt['cantus'], 'cantus')
for x in final_cpt['cantus']:  # cf
    c_note = notes[x % 12]
    c_octave = ''
    for a, b in octaves.items():
        if x in b:
            c_octave = a
    cantus.append(note.Note(c_note + c_octave, type='whole'))

for x in final_cpt['counterpoint']:  # ctpt
    c_note = notes[x % 12]
    c_octave = ''
    for a, b in octaves.items():
        if x in b:
            c_octave = a
    counterpoint.append(note.Note(c_note + c_octave, type='whole'))

S = stream.Score()
S.insert(counterpoint)
S.insert(cantus)
print('finished in', time() - st, 'seconds')
S.show()
```
